```
sign = pd.read_csv('data/signnames.csv')
sign.head()
```

```
dict_signs=sign.to_dict()['b']
```

```
X_train[y_train==10].shape #ile jest znaków z labelem o numerze 10
```

```

for id_sign in dict_signs.keys():
  
  given_signs = X_train[y_train==id_sign]
  plt.figure(figsize=(15,5))
  for i in range(9): #9 znaków w rzędzie
    plt.subplot('19{0}'.format(i+1))
    plt.imshow(given_signs[i])
    plt.axis('off')
plt.show()
plt.tight_layout()
```

```
cnt=Counter(y_train).most_common()
cnt
```

```
id_labels, cnt_labels = zip(*cnt) #utworzenie dwóch list

ids=range(len(id_labels))

plt.figure(figsize=(15,5))
plt.bar(ids, cnt_labels)

plt.xlabel('znaki')
labels=[dict_signs[id_labels[id_]] for id_ in id_labels]
plt.xticks(ids, labels, rotation='vertical')
plt.title('Znaki drogowe - liczba wystąpień wg znaku')
plt.show()
```
